```python
import numpy as np
import matplotlib.pyplot as plt

#A2
# H = 4.84
# L = 2.76
#A3
# H = 4.92
# L = 3.29
# #A1 
# H = 4.60
# L = 2.28
# #A4
# H = 4.91
# L = 4.50
#in these tests R2 = 50K_ohm
x = np.logspace(1,4.5,5000)

def findR1(VC):
    #R1 = R2(VDD-VC)/VC
    result = 50000*(5.0-VC)/VC
    return result

def clacVoltageCurve(R1, R2Array):
    R1Array = R1*np.ones(len(R2Array))
    voltageVals = 5*R2Array/(R1Array+R2Array)
    return voltageVals
#This is a notebook for using the voltage devider formula to detmine the value of an ideal resistor
'''       
          VDD (5V)  ------------+ 
                                |
                              <R1>
                                |
                               -+-----------> Vc
                                |
                              <R2>
                                |
                                +------------------+
                                                   |
                                                  ===
                                                  Gnd
Vc = VDD*(R2)/(R1+R2)
Vc(R1+R2) = VDD*(R2)
VcR1 = VDD*R2 - VcR2
R1 = R2(VDD-VC)/VC

'''
```

```python

#A2
H = 4.84
L = 2.76

print('H R1=',round(findR1(H)))
print('L R1=',round(findR1(L)))

y1 = clacVoltageCurve(findR1(H),x)
y2 = clacVoltageCurve(findR1(L),x)
y3 = y1 - y2
print('\nOptimal A2 R2=',round(x[np.argmax(y3)]),' Ohms')
plt.plot(x,y3)
plt.scatter(x[np.argmax(y3)],np.max(y3),c='r',marker='x')
# print(np.max(y3))
# print(y3[np.argmax(y3)])
# print(np.argmax(y3))
```
